                                                               Top 10 Artists

```python
top_artists = df['artists'].value_counts().head(10)

print(top_artists)
```

```python
plt.style.use("dark_background")

plt.figure(figsize=(12,6))

top_artists.plot(kind='bar')

plt.title("Top 10 Most Popular Artists", fontsize=16)
plt.xlabel("Artists")
plt.ylabel("Tracks")

plt.xticks(rotation=45)

plt.show()
```

Insights:
- Drake and Taylor Swift dominate the dataset.
- Highly active artists tend to maintain stronger visibility on Spotify.
- Popular artists release tracks more consistently across years.

                                                         Most Streamed Songs

```python
top_songs = df.sort_values(by='popularity', ascending=False).head(10)

plt.style.use("dark_background")

plt.figure(figsize=(12,6))

sns.barplot(
    x='popularity',
    y='track_name',
    data=top_songs
)

plt.title("Top 10 Most Popular Songs", fontsize=16)

plt.xlabel("Popularity")
plt.ylabel("Songs")

plt.show()
```

- The top tracks dominate Spotify engagement.
- Popular songs usually belong to globally recognized artists.
- High popularity often correlates with strong streaming performance.

                                                                  Popularity Distribution

```python
plt.figure(figsize=(10,6))

sns.histplot(df['popularity'], bins=30)

plt.title("Popularity Distribution")

plt.xlabel("Popularity Score")
plt.ylabel("Number of Songs")

plt.show()
```

- Most songs fall into medium popularity ranges.
- Only a small percentage achieve extremely high popularity.
- Spotify popularity distribution is highly competitive.

                                                              Danceability Analysis

```python
plt.figure(figsize=(10,6))

sns.histplot(df['danceability'], bins=30)

plt.title("Danceability Distribution")

plt.xlabel("Danceability")
plt.ylabel("Song Count")

plt.show()
```

- Most tracks are moderately danceable.
- Highly danceable songs tend to dominate trending playlists.
- Spotify favors energetic and engaging tracks.

                                                                  Correlation Heatmap

```python
numeric_df = df.select_dtypes(include=['float64', 'int64'])

plt.figure(figsize=(14,10))

sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap='coolwarm'
)

plt.title("Feature Correlation Heatmap")

plt.show()
```

                                                           Explicit vs Non-Explicit Songs

```python
explicit_counts = df['explicit'].value_counts()

plt.figure(figsize=(6,6))

plt.pie(
    explicit_counts,
    labels=['Non-Explicit', 'Explicit'],
    autopct='%1.1f%%'
)

plt.title("Explicit vs Non-Explicit Songs")

plt.show()
```

- Non-explicit songs generally dominate the platform.
- Explicit content still forms a significant portion of modern music.
- Listener demographics influence content trends.
